```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# Load data
df = pd.read_csv("/content/drive/MyDrive/archive (1)/loan_data.csv")

X = df.drop("loan_status", axis=1)
y = df["loan_status"]

num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
])

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    random_state=42
)

pipeline = Pipeline([
    ("prep", preprocessor),
    ("model", model)
])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

pipeline.fit(X_train, y_train)

print(classification_report(y_test, pipeline.predict(X_test)))

# Save model
joblib.dump(pipeline, "model.pkl")
joblib.dump(preprocessor, "preprocessor.pkl")
```

```python
import shap
import numpy as np

explainer = shap.TreeExplainer(pipeline.named_steps["model"])
X_transformed = pipeline.named_steps["prep"].transform(X_train)

shap_values = explainer.shap_values(X_transformed)

# Top 5 global features
# Corrected indexing: Access the SHAP values for the second class across all samples and features.
shap.summary_plot(shap_values[:, :, 1], X_transformed, max_display=5)
```

```python
def approval_rate(df, preds, group):
    df_temp = df.copy()
    df_temp["prediction"] = preds
    return df_temp.groupby(group)["prediction"].mean()

y_pred = pipeline.predict(X_test)

print("Gender fairness:")
print(approval_rate(X_test, y_pred, "person_gender"))

print("\nEducation fairness:")
print(approval_rate(X_test, y_pred, "person_education"))
```

```python
from fastapi import FastAPI
import pandas as pd
import joblib
import shap

app = FastAPI()

pipeline = joblib.load("model.pkl")
preprocessor = pipeline.named_steps["prep"]
model = pipeline.named_steps["model"]

explainer = shap.TreeExplainer(model)

@app.post("/score")
def score(data: dict):
    X = pd.DataFrame([data])

    prob = pipeline.predict_proba(X)[0][1]
    pred = int(prob >= 0.5)

    X_trans = preprocessor.transform(X)
    shap_vals = explainer.shap_values(X_trans)[1][0]

    features = preprocessor.get_feature_names_out()
    top5 = sorted(
        zip(features, shap_vals),
        key=lambda x: abs(x[1]),
        reverse=True
    )[:5]

    return {
        "prediction": pred,
        "approval_probability": round(float(prob), 3),
        "top_5_shap_features": [
            {"feature": f, "impact": round(float(v), 4)}
            for f, v in top5
        ]
    }
```
